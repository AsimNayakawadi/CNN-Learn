```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ============================================
# STEP 1: Load and handle the BGR gotcha
# ============================================
img_bgr = cv2.imread("photo.jpg")          # OpenCV loads as BGR!
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  # Convert for display

print(f"Shape: {img_rgb.shape}")           # (H, W, 3)
print(f"Dtype: {img_rgb.dtype}")           # uint8
print(f"Range: [{img_rgb.min()}, {img_rgb.max()}]")  # [0, 255]

# ============================================
# STEP 2: Convert to different color spaces
# ============================================
img_hsv  = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

# Split into individual channels
r, g, b = img_rgb[:,:,0], img_rgb[:,:,1], img_rgb[:,:,2]
h, s, v = img_hsv[:,:,0], img_hsv[:,:,1], img_hsv[:,:,2]

# ============================================
# STEP 3: Visualize everything
# ============================================
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
fig.suptitle("Color Space Channel Breakdown", fontsize=16, fontweight='bold')

# Row 1: Original + RGB channels
axes[0,0].imshow(img_rgb)
axes[0,0].set_title("Original (RGB)")

axes[0,1].imshow(r, cmap='Reds')
axes[0,1].set_title("R channel")

axes[0,2].imshow(g, cmap='Greens')
axes[0,2].set_title("G channel")

axes[0,3].imshow(b, cmap='Blues')
axes[0,3].set_title("B channel")

# Row 2: HSV channels (with colormaps that match meaning)
axes[1,0].imshow(img_rgb)
axes[1,0].set_title("Original")

axes[1,1].imshow(h, cmap='hsv')           # Rainbow colormap for Hue
axes[1,1].set_title("H (Hue) — which color")

axes[1,2].imshow(s, cmap='gray')
axes[1,2].set_title("S (Saturation) — how vivid")

axes[1,3].imshow(v, cmap='gray')
axes[1,3].set_title("V (Value) — how bright")

# Row 3: Grayscale + channel comparisons
axes[2,0].imshow(img_gray, cmap='gray')
axes[2,0].set_title("Grayscale")

# Show what happens when you zero out channels
red_only = img_rgb.copy(); red_only[:,:,1] = 0; red_only[:,:,2] = 0
axes[2,1].imshow(red_only)
axes[2,1].set_title("Red channel only (color)")

green_only = img_rgb.copy(); green_only[:,:,0] = 0; green_only[:,:,2] = 0
axes[2,2].imshow(green_only)
axes[2,2].set_title("Green channel only (color)")

blue_only = img_rgb.copy(); blue_only[:,:,0] = 0; blue_only[:,:,1] = 0
axes[2,3].imshow(blue_only)
axes[2,3].set_title("Blue channel only (color)")

for ax in axes.flat:
    ax.axis('off')

plt.tight_layout()
plt.savefig("color_channels.png", dpi=150, bbox_inches='tight')
plt.show()
```
