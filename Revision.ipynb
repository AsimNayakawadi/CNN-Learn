### Problem 1 — List comprehension

Make a new list with only words that have more than 3 letters.

```python
words = ["hello", "world", "python", "is", "fun"]
```

```python
res = [f for f in words if len(f) > 3]
print(res)
```

## Problem 2 — List comprehension

Make a new list where every number is multiplied by 2, but only if the number is greater than 6.

```python
numbers = [3, 7, 12, 5, 18, 21, 4, 9]
```

```python
res2 = [n*2 for n in numbers if n>6]
print(res2)
```

## Problem 3 — List comprehension

Keep only files that start with "img" and replace .png with ".jpg" for each one.

```python
files = ["img_001.png", "mask_001.png", "img_002.png", "mask_002.png", "img_003.png"]
```

```python
res3 = [ f.replace(".png",".jpg") for f in files if f.startswith("img")]
print(res3)
```
