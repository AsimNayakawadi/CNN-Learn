## Lets build a convolution

#### Block 1: A convolution layer in PyTorch

```python
import torch 
import torch.nn as nn


conv = nn.Conv2d(in_channels=3, out_channels=16, kernel_size = 3, padding = 1)
print(conv.weight.shape)
print(conv.bias.shape)
```

### Block 2: Apply it to a fake batch

```python
x = torch.randn(1, 3, 256, 256) # batch of 1, 3 channels, 256x256

y = conv(x)
print(y.shape)
```

### Block 3: A downsampling convolution

```python
conv_down = nn.Conv2d(16, 32, kernel_size=3, stride=2, padding=1)
y2 = conv_down(y)
print(y2.shape)   
```

### Block 4: A pooling layer (no learned weights, just downsamples)

```python
pool = nn.MaxPool2d(kernel_size=2, stride=2)
y3 = pool(y2)
print(y3.shape)    
```
